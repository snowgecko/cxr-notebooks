```python
import pandas as pd
from functools import reduce
```

```python
root = '/well/papiez/users/hri611/python/foundation-models-radiology/'
biomedclip_df = pd.read_csv(root+"ptx_biomedclip_scores.csv")
biomedclip_df['image_paths'] = biomedclip_df['image_paths'].apply(lambda x: x[:-4]) # remove .png/dicom

biomedgpt_df = pd.read_csv(root+"ptx_biomedgpt_scores.csv")
biomedgpt_df = biomedgpt_df.rename(columns={"image_path": "image_paths"})
biomedgpt_df['image_paths'] = biomedgpt_df['image_paths'].apply(lambda x: x.replace(('/well/papiez/users/hri611/python/foundation-models-radiology/PTX Head to Head Study Data/'), ''))
biomedgpt_df['image_paths'] = biomedgpt_df['image_paths'].apply(lambda x: x[:-4]) # remove .png/dicom

google_df = pd.read_csv(root+"ptx_google_scores.csv")
google_df['image_paths'] = google_df['image_paths'].apply(lambda x: x[:-4]) # remove .png/dicom

chexagent_df = pd.read_csv(root+"ptx_chexagent_scores.csv")
chexagent_df = chexagent_df.rename(columns={"image_path": "image_paths"})
chexagent_df['image_paths'] = chexagent_df['image_paths'].apply(lambda x: x.replace(('/well/papiez/users/hri611/python/foundation-models-radiology/PTX Head to Head Study Data/'), ''))
chexagent_df['image_paths'] = chexagent_df['image_paths'].apply(lambda x: x[:-4]) # remove .png/dicom

radfm_df = pd.read_csv(root+"ptx_radfm_scores.csv")
radfm_df['image_paths'] = radfm_df['image_paths'].apply(lambda x: x[:-4]) # remove .png/dicom
radfm_df.drop(columns=['Unnamed: 0'], inplace=True)

medgemma_df = pd.read_csv(root+"ptx_medgemma_scores.csv")
medgemma_df['image_paths'] = medgemma_df['image_path'].apply(lambda x: x.replace(('/well/papiez/users/hri611/python/foundation-models-radiology/PTX Head to Head Study Data/'), ''))
medgemma_df['image_paths'] = medgemma_df['image_paths'].apply(lambda x: x[:-4]) # remove .png/dicom
medgemma_df.drop(columns=['image_path'], inplace=True)
```

```python
biomedclip_df = biomedclip_df.rename(columns=lambda x: f"biomedclip_{x}" if x != "image_paths" else x)
biomedgpt_df = biomedgpt_df.rename(columns=lambda x: f"biomedgpt_{x}" if x != "image_paths" else x)
google_df = google_df.rename(columns=lambda x: f"cxrfoundation_{x}" if x != "image_paths" else x)
chexagent_df = chexagent_df.rename(columns=lambda x: f"chexagent_{x}" if x != "image_paths" else x)
radfm_df = radfm_df.rename(columns=lambda x: f"radfm_{x}" if x != "image_paths" else x)
medgemma_df = medgemma_df.rename(columns=lambda x: f"medgemma_{x}" if x != "image_paths" else x)

# Merge all on 'image_paths'
dfs = [google_df,biomedclip_df,chexagent_df, biomedgpt_df, radfm_df, medgemma_df]
merged_df = reduce(lambda left, right: pd.merge(left, right, on='image_paths', how='outer'), dfs)

merged_df.drop(columns=['biomedgpt_response'], inplace=True)
```

```python
merged_df.to_csv('ptx_6_scores.csv')
```
